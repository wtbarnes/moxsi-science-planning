```python
import roentgen
import roentgen.absorption
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import quantity_support
```

```python
alloy = roentgen.absorption.Material(
    {'Cu': 0.25, 'W': 0.75},
    thickness=350*u.micron,
    density=14.5*u.g/(u.cm**3),
)
```

```python
energy = np.linspace(1,400,1000) * u.keV
```

```python
with quantity_support():
    plt.plot(energy, alloy.absorption(energy))
```
